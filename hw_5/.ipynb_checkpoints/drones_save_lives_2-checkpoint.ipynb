```python
import numpy as np
import pylab as plt
import skfmm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
def plot_surface(surface):
    fig = plt.figure(figsize=(15, 10))
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, surface, cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
```

```python
n = 200
l = 20

X, Y = np.meshgrid(np.linspace(-l/2, l/2, n), np.linspace(-l/2, l/2, n))
phi = np.ones((n, n))
phi[(X==0) & (Y==0) | ((X==l/2) & (Y==l/2))] = -1

mask = np.less_equal(np.sqrt(X**2 + Y**2), 5)
phi = np.ma.MaskedArray(phi, mask)
distance = skfmm.distance(phi, dx=1)
```

```python
plt.figure(figsize=(10, 8))
plt.contour(X, Y, phi, [0], linewidths=(3), colors='black')
plt.contour(X, Y, phi.mask, [0], linewidths=(3), colors='red')
plt.contour(X, Y, distance, 50)
plt.colorbar()
plt.show()
```

```python
def convert_distance(distance):
    distance2 = np.copy(distance)
    for i in range(n):
        for j in range(n):
            if type(distance[i, j]) is not np.float64:
                distance2[i, j] = 1e3
    return distance2
```

```python
def min_path(distance, m, n, size):
    distance = convert_distance(distance)
    travel = []
    total_cost = np.zeros((size, size))
    total_cost[0, 0] = distance[0, 0]
 
    for i in range(1, m+1):
        total_cost[i, 0] = total_cost[i-1, 0] + distance[i, 0]
    for j in range(1, n+1):
        total_cost[0, j] = total_cost[0, j-1] + distance[0, j]
        
    for i in range(1, m+1):
        for j in range(1, n+1):
            total_cost[i, j] = min(total_cost[i-1, j-1], total_cost[i-1, j], total_cost[i, j-1]) + distance[i, j]
 
    return total_cost
```

```python
total_cost = min_path(distance, 199, 199, n)
```

```python
plt.figure(figsize=(10, 8))
plt.title('Travel time from the boundary with an obstacle')
plt.contour(X, Y, total_cost, 400)
plt.colorbar()
plt.savefig('2d_phi_travel_time_mask.png')
plt.show()
```

```python
plot_surface(total_cost)
plot_surface(convert_distance(distance))
```

```python
def shortest_path(gradient_norm, start_point):
    next_point = []
    next_point.append(start_point)
    travel = []
    while next_point[-1][0] > 0 or next_point[-1][0] > 0:
        min_cost = 1000
        best_point = (0, 0)
        for (i, j) in [(-1, -1), (-1, 0), (0, -1)]:
            cost = gradient_norm[next_point[-1][0] + i, next_point[-1][1] + j]
            if cost < min_cost:
                best_point = (next_point[-1][0] + i, next_point[-1][1] + j)
                min_cost = cost
        next_point.append(best_point)
        travel.append((best_point[0]/10 - 10, best_point[1]/10 - 10))
    return travel
```

```python
travel = shortest_path(distance, (199, 199))
```

```python
gradient = np.gradient(distance)
gradient_norm = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        gradient_norm[i, j] = np.linalg.norm([gradient[0][i, j], gradient[1][i, j]])
        
plt.figure(figsize=(10, 8))
plt.contour(X, Y, phi, [0], linewidths=(3), colors='black')
plt.contour(X, Y, phi.mask, [0], linewidths=(3), colors='red')
plt.contour(X, Y, distance, 50)
plt.contour(X, Y, gradient_norm, 10)
plt.plot(*zip(*travel))
plt.colorbar()
plt.show()
```
